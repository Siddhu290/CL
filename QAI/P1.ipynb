```python
# python -m venv qiskit_env
 
#  qiskit_env\Scripts\activate  

# pip install qiskit 
# pip install jupyter
# pip install qiskit-aer  
# pip install matplotlib 
# pip install qiskit-ibm-runtime  



# Step 1: Initialise the quantum and classical registers
# The first step is to initialise a 16 qubit register . This is done by the following code:

# q = QuantumRegister(16,'q')

# Next we initialise the 16 bit classical register with the following code:

# c = ClassicalRegister(16,'c')

# Step 2: Create the circuit 
# Next we create quantum circuit using the following code:

# circuit = QuantumCircuit(q,c)

# Step 3: Apply a Hadamard gate to all qubits 
# Then we need to apply a Hadamard gate. This gate is used to put a qubit in to a superposition of 1 and 0 such that when we measure the qubit it will be 1 or a 0 with equal probability. 

# This is done with the following code:

# circuit.h(q)

# Step 4: Measure the qubits
# After this we measure the qubits. This measurement will collapse the qubits superposition in to either a 1 or a 0. 

# This is done with the following code: 

# circuit.measure(q,c) 
# Import necessary libraries from Qiskit




# QuantumCircuit is used to create a quantum circuit
from qiskit import QuantumCircuit  
# plot_histogram is used for visualizing measurement results as histograms
from qiskit.visualization import plot_histogram, circuit_drawer  
# Aer provides simulation backends like the qasm_simulator
from qiskit_aer import Aer  

# Step 1: Create a Quantum Circuit with 16 qubits and 16 classical bits
qc = QuantumCircuit(16, 16)  # This creates a quantum circuit with 16 qubits and 16 classical bits

# Step 2: Apply the Hadamard gate to all 16 qubits
qc.h(range(16))  
# The Hadamard gate (h) puts each qubit into a superposition, 
# creating an equal probability of being measured in either the 0 or 1 state. 
# This is what introduces the randomness in quantum circuits.

# Step 3: Measure all the qubits
qc.measure(range(16), range(16))  
# This command measures all 16 qubits and stores the results in the corresponding 16 classical bits.

# Step 4: Print the quantum circuit
print(qc)  
# The print statement shows the structure of the quantum circuit (gates and measurements applied).

# Step 5: Visualize the quantum circuit
circuit_drawer(qc)  
# This will generate a graphical representation of the circuit, showing the qubits and operations applied.

# Step 6: Set up the simulator backend (Aer is Qiskit's high-performance simulator)
simulator = Aer.get_backend('qasm_simulator')  
# The qasm_simulator simulates the quantum circuit on a classical computer using a probabilistic model.
# It's useful when you want to simulate the results of quantum measurements.

# Step 7: Execute the circuit using the simulator, with 1 shot (1 execution)
job = simulator.run(qc, shots=1)  
# A "shot" refers to one run of the quantum circuit. We are using one shot to get a single result from the quantum circuit.

# Step 8: Get the result of the simulation
result = job.result()  
# The result method fetches the outcome of the simulation, including measurement counts and other relevant data.

# Step 9: Extract the measurement counts
counts = result.get_counts(qc)  
# The get_counts method returns a dictionary containing the counts of the measurement outcomes for each possible state (e.g., 000000... or 111111...).

# Step 10: Convert the result into a binary number
key = list(counts.keys())[0]
random_number = int(key,2) # binary to decimal integer

# The first key in the counts dictionary is the measurement outcome in binary form (as a string). 
# We convert this binary string into a decimal integer using `int(binary_string, 2)`.

# Step 11: Format the binary result into a 16-bit string
binary_number = bin(random_number)[2:].zfill(16)  
# The `bin()` function converts the random_number into a binary string.
# We use slicing `[2:]` to remove the '0b' prefix that Python adds to binary numbers.
# `zfill(16)` pads the binary number with leading zeros to ensure it's 16 bits long.

# Step 12: Print the results
print("Random number: ", random_number)  
# The decimal representation of the random number is printed.

print("Binary number: ", binary_number)  
# The binary representation of the random number is printed.
```
